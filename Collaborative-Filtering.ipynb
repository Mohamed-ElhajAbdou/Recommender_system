### User-User Collaborative-Filtering

```python
def train():
    df = pd.read_csv('rating.csv')
    df = df.pivot(index='userId',columns='movieId',values='rating')
    df.fillna(0,inplace=True)
    cor_matrix = df.iloc[:100,:100].T.corr()
    return cor_matrix
    
def get_dif(row):
    if row['from']>=3 and row['to']==0:
        return row['from']
    else:
        return 0
    
def predict(user_id,n=25):
    from_id = cor_matrix[user_id].sort_values(ascending=False).index[1]
    rec = pd.DataFrame({
        'from':df.loc[from_id],
        'to':df.loc[user_id]
    })
    rec['diff'] =rec.apply(get_dif,axis=1)
    rec = rec.sort_values(by='diff',ascending=False)
    return rec.index[:n]

```

```python
cor_matrix = train()
```

```python
predict(2,5)
```

```python
predict(2)
```

### Item-Item Collaborative-Filtering

```python
def train():
    df = pd.read_csv('rating.csv')
    df = df.pivot(index='userId',columns='movieId',values='rating')
    df.fillna(0,inplace=True)
    cor_matrix = df.iloc[:100,:100].corr()
    return cor_matrix

def predict(movie_id,n=25):
    return cor_matrix[movie_id].sort_values(ascending=False).index[1:n+1]
    
```

## here im working on the movies

```python
cor_matrix = train()
predict(1,10)
```

```python
predict(1,10)
```
