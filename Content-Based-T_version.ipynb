## Content-Based in only one function

```python
import re
from sklearn.feature_extraction.text import CountVectorizer
from nltk.stem import ISRIStemmer
from sklearn.neighbors import NearestNeighbors
```

```python
cv = CountVectorizer()
ps = ISRIStemmer()   ## arabic
#sw = open('sw.txt','r').read().split('\n')   ## arabic

def clean(text):
    text = str(text)
    text = re.sub('[^ء-ي0-9 ]+','',text).strip()  ## arabic
    filtered_words = []
    for word in text.split():   ## arabic 
        #if word not in sw:
        filtered_words.append(ps.stem(word))
    return ' '.join(filtered_words)

def train(n=5):
    df = pd.read_csv('articles.csv')
    df = df[['url', 'title', 'creator', 'keywords','article']]
    df.dropna(inplace=True)
    df['text'] = df.title+" "+df.creator+" "+df.keywords+" "+df.article
    df['clean_text'] = df.text.apply(clean)
    x  = cv.fit_transform(df.clean_text).toarray()
    nbrs = NearestNeighbors(n_neighbors=n+1)
    nbrs.fit(x)
    distances, indices = nbrs.kneighbors(x)
    columns = []
    for i in range(1,n+1):
        columns.append('rec'+str(i))
    columns.insert(0,'url')
    rec = pd.DataFrame(indices,columns=columns)
    rec = rec.applymap(lambda x: df.iloc[x].url)
    return rec

def predict(url):
    return rec[rec.url==url].values[0][1:]    
```

```python
rec = train(10)
```
